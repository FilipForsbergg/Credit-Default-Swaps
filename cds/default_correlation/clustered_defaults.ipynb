```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from cds.data.build_portfolio import load_all_series_cdx_hy_us_data
```

## Load all CDX HY US data

```python
excel_file = load_all_series_cdx_hy_us_data()

for sheet_name in excel_file.sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet_name)
    df.columns = ['Empty', 'Company', 'Weight', 'Currency', 'Seniority', 'RED_Code', 'SP_ID']
    df = df[1:].drop(columns=['Empty'])
    df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
    
```

```python
df = pd.read_excel('CDX_HY_Defaults.xlsx')
df['Default_Date'] = pd.to_datetime(df['Default_Date'], format='%Y-%m-%d')
counts = df.set_index('Default_Date').resample('QE').size()
counts.index = counts.index.to_period('Q')

counts.plot(kind='bar', figsize=(12,6), color='firebrick')
plt.title('Defaults per quarter in CDX HY')
plt.ylabel('No. of Defaults')
plt.xlabel('Quarter')
plt.show()
```

```python
default_map = pd.Series(df.Default_Date.values, index=df.RED_Code).to_dict()
def get_series_default_path(series_df, series_name):
    series_codes = series_df['RED_Code'].unique()
    dates = [default_map[code] for code in series_codes if code in default_map]
    dates.sort()

    path = pd.DataFrame({'Date': dates})
    path['Cumulative_Defaults'] = range(1, len(dates) + 1)

    return path
```
